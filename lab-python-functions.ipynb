```python
   
def initialize_inventory(products): # products es lista
    inventory = {}
    for product in products:
        quantity = int(input(f"Introduce la cantidad disponible de {product}: "))
        inventory[product] = quantity
    return inventory #diccionario


```

```python
def get_customer_orders():
    customer_orders = set()
    sigue_pedido = input("¿Vas a pedir algo yes/no: ").lower()
    
    while sigue_pedido == "yes":
        producto = input("Dime qué quieres pedir: ")
        customer_orders.add(producto)
        sigue_pedido = input("¿Vas a pedir algo yes/no: ").lower()


    return customer_orders
```

```python
def update_inventory(customer_orders ,inventory):
    for producto in customer_orders:
        if producto in inventory:
            quantity = int(input(f"Cuantos quieres pedir de {producto}"))
            inventory[producto] -= quantity
    return inventory
```

```python
def calculate_order_statistics(customer_orders, products):
    total_products_ordered = len(customer_orders)
    percentage_unique = round((total_products_ordered / len(products)) * 100, 2)
    return total_products_ordered, percentage_unique

```

```python
def print_order_statistics(order_statistics):
    print(f"total_products_ordered : {order_statistics[0]}")
    print(f"porcentage_of_unique_products_ordered: {order_statistics[1]}%")

```

```python
def print_updated_inventory(inventory):
    print("\nInventario actualizado:")
    for product, quantity in inventory.items():
        print(f"{product}: {quantity}")

```

```python
lista_productos = ['manzanas', 'peras', 'melones', 'sandias']

inventory = initialize_inventory(lista_productos)
customer_orders = get_customer_orders()
inventory = update_inventory(customer_orders, inventory)

order_statistics = calculate_order_statistics(customer_orders, lista_productos)

print_order_statistics(order_statistics)
print_updated_inventory(inventory)
```
