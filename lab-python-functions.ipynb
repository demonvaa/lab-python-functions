1. Define a function named `initialize_inventory` that takes `products` as a parameter. Inside the function, implement the code for initializing the inventory dictionary using a loop and user input.

```python
   
def initialize_inventory(products):
    inventory = {}
    for product in products:
        quantity = int(input(f"Introduce la cantidad disponible de {product}: "))
        inventory[product] = quantity
    return inventory
```

2. Define a function named `get_customer_orders` that takes no parameters. Inside the function, implement the code for prompting the user to enter the product names (por ejemplo me va a quitar 5 unidades) using a loop. The function should return the `customer_orders` set. 

```python
def get_customer_orders():
    customer_orders = set()
    pedido_completado = "no"

    while pedido_completado != "yes":
        producto = input("Dime qué quieres pedir: ")
        customer_orders.add(producto)

        pedido_completado = input("¿Has terminado el pedido? yes/no: ").lower()

    return customer_orders

```

3. Define a function named `update_inventory` that takes `customer_orders` and `inventory` as parameters. Inside the function, implement the code for updating the inventory dictionary based on the customer orders.

```python
def update_inventory(customer_orders ,inventory):
    for producto in customer_orders:
        if producto in inventory:
            quantity = int(f"Dime cuantos quieres pedir de {producto}")
            inventory[producto] -= quantity
    return inventory
```

4. Define a function named `calculate_order_statistics` that takes `customer_orders` and `products` as parameters. Inside the function, implement the code for calculating the order statistics (total products ordered, and percentage of unique products ordered). The function should return these values.

```python
def calculate_order_statistics(customer_orders,products):
    total_products_ordered = len(customer_orders)
    quantity_product = len(products)
    percentage_of_unique_products_ordered = round((total_products_ordered /quantity_product)*100,2)
    return total_products_ordered, quantity_product, percentage_of_unique_products_ordered
```
